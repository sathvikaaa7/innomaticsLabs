```python
#2. Pass/Fail Analyzer
marks =  [45, 78, 90, 33, 60]
fail_count = 0
pass_count = 0
for i in marks:
  if i >= 50:
    pass_count += 1
  else:
    fail_count += 1
print("Passed students: ", pass_count)
print("Failed Students: ", fail_count)
```

```python
#3. Simple Data Cleaner
names = [" Alice ", "bob", " CHARLIE "]
x = []
for i in names:
  x.append(i.strip().lower())
print(x)
```

```python
#4.Message Length Analyzer

messages = ["Hi", "Welcome to the platform", "OK"]
for i in messages:
  x = len(i)
  print(f"Message: '{i}' it's Length is: {x}")

  if x > 10:
    print(f"The message '{i}''s lenght is more than 10")
```

```python
#5.  Error Message Detector
logs = ["INFO", "ERROR", "WARNING", "ERROR"]
count = 0
for i in logs:
  if i == "ERROR":
    count += 1
print("Total error count = ",count)
```
